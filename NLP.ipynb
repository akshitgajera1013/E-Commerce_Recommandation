```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
laptops = pd.read_csv("flipkart_laptops.csv", encoding='latin1')
mobiles = pd.read_csv("flipkart_mobiles.csv", encoding='latin1')
tv = pd.read_csv("flipkart_tv.csv", encoding='latin1')
fridge = pd.read_csv("flipkart_refrigerator.csv", encoding='latin1')
watch = pd.read_csv("flipkart_smart_watch.csv", encoding='latin1')
wm = pd.read_csv("flipkart_washing_machine.csv", encoding='latin1')
```

```python
laptops["category"] = "laptop"
mobiles["category"] = "mobile"
tv["category"] = "tv"
fridge["category"] = "refrigerator"
watch["category"] = "smart_watch"
wm["category"] = "washing_machine"
```

```python
df = pd.concat([laptops, mobiles, tv, fridge, watch, wm], ignore_index=True)
```

```python
df=df[['Name','Brand','category','Details','Selling Price','Ratings','No_of_ratings']]
```

```python
df=df.drop_duplicates().reset_index(drop=True)
```

```python
df['combined']=df['Name']+df['Brand']+df['category']+df['Details']
```

```python
stop_words=set(nltk.corpus.stopwords.words('english'))
```

```python
def preprocess(text):
    text=text.lower()
    tokens=nltk.word_tokenize(text)
    tokens=[token for token in tokens if token.isalpha()]
    tokens=[token for token in tokens if token not in stop_words]
    lemmatizer=nltk.stem.WordNetLemmatizer()
    tokens=[lemmatizer.lemmatize(token) for token in tokens]
    preprocessed_text=' '.join(tokens)
    return preprocessed_text
```

```python
df['combined']=df['combined'].apply(preprocess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=5000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['combined'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommand(product,n=10):
    product_index=df[df['Name']==product].index[0]
    similarity_scores=cosine_similarity(tfidf_matrix[product_index],tfidf_matrix).flatten()
    # Get indices of products sorted by similarity scores in descending order
    recommended_products_indices=similarity_scores.argsort()[::-1]

    # Exclude the input product itself from recommendations
    recommended_products_indices=recommended_products_indices[1:]

    # Get the top n recommended product names
    recommended_products = df['Name'].iloc[recommended_products_indices[:n]].tolist()

    return recommended_products
```

```python
sample_laptop_name = df[df['category'] == 'laptop']['Name'].iloc[0]
print(f"Using sample laptop name: {sample_laptop_name}")
recommand(sample_laptop_name)
```
